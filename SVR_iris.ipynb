```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
y_regression = X[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_regression,test_size = 0.2, random_state = 42)
```

```python
svr = SVR(kernel = 'sigmoid') #linear #rbf #sigmoid #tanh
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sepal Length")
plt.ylabel("Predicted Sepal Length")
plt.title("SVR: Actual vs Predicted Sepal Length")
plt.show()
```
