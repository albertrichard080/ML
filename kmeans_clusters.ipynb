```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
#load the iris datasets
iris = datasets.load_iris()
X = iris.data
```

```python
#define the no of clusters
k=3
```

```python
#create a KMeans model
Kmeans = KMeans(n_clusters=k, random_state=0)
```

```python
# fit the model
Kmeans.fit(X)
```

```python
# get the cluter assignments for the each data point
labels = Kmeans.labels_
```

```python
#get the coordinates of the clustere centers
centers = Kmeans.cluster_centers_
```

```python
#visualize the clusters
plt.scatter(X[labels==0][:, 0],X[labels==0][:, 1], label='Cluster 1', c='red')
plt.scatter(X[labels==1][:, 0],X[labels==1][:, 1], label='Cluster 2', c='green')
plt.scatter(X[labels==2][:, 0],X[labels==2][:, 1], label='Cluster 3', c='blue')
plt.scatter(centers[:, 0], centers[:, 1], marker = '*', s=100, linewidths=3, color='black', label='centroids')
plt.legend()
```
